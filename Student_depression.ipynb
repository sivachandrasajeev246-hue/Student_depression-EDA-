```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Doc Scanner/Student Depression Dataset.csv")
df.head()
```

```python
df.describe(include="all")
```

```python
df.info()
```

```python
print(df.shape)
print(df.isnull().sum())
print(df.dtypes.value_counts())
print(df.nunique())
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("\n Summary of Categorical Columns:")
print(df[categorical_cols].describe())
import numpy as np
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"\nNumerical Columns: {numerical_cols}")

print(f"\nCategorical Columns: {list(categorical_cols)}")
```

```python
#dropping 3 columns where financial stress has Nan value,since it is big data no problem removing 3 rows
df = df.dropna(subset=['Financial Stress'])
```

```python
df = df.drop(columns=['id'])
```

```python
duplicated_count=df.duplicated().sum()
print(duplicated_count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df_num=df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,8))
sns.heatmap(df_num.corr(),annot=True,cmap='coolwarm')
```

```python
#T o know about class distribution
sns.countplot(x='Depression', data=df)
plt.title("Depression Label Count")
plt.xlabel("Depression (0 = No, 1 = Yes)")
plt.ylabel("Number of Students")
plt.show()
```

```python
#how stress level affect depression
plt.figure(figsize=(10, 6))
sns.boxplot(x='Depression', y='Financial Stress', data=df)
plt.title('Financial Stress vs Depression')
plt.xlabel("Depression (0 = No, 1 = Yes)")
plt.ylabel("Financial Stress ")
plt.show()
```

```python
sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="Gender", hue="Depression", data=df, palette="dark")
plt.title("Depression Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x="Depression", y="CGPA", data=df, palette="muted")
plt.title("CGPA vs Depression")
plt.xlabel("Depression")
plt.ylabel("CGPA")
plt.show()
```

```python
sns.countplot(x="Sleep Duration", hue="Depression", data=df, palette="deep")
plt.title("Depression by Sleep Duration")
plt.xlabel("Sleep Duration")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="Have you ever had suicidal thoughts ?", hue="Depression", data=df, palette="coolwarm")
plt.title("Suicidal Thoughts vs Depression")
plt.xlabel("Suicidal Thoughts")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
